```python
a=[[6, -8, 3]]

b=[[1], [1], [4]]


m = len(a)                                            # a: m × n
n = len(b)                                            # b: n × k
k = len(b[0])

c = [[None for __ in range(k)] for __ in range(m)]    # c: m × k

for i in range(m):
    for j in range(k):
        c[i][j] = sum(a[i][kk] * b[kk][j] for kk in range(n))

print(c)
```

```python
a=[[-3, -7], [-10, 6], [0, -6], [-1, -3]]

b=[[-8, 9], [4, -2]]


m = len(a)                                            # a: m × n
n = len(b)                                            # b: n × k
k = len(b[0])

c = [[None for __ in range(k)] for __ in range(m)]    # c: m × k

for i in range(m):
    for j in range(k):
        c[i][j] = sum(a[i][kk] * b[kk][j] for kk in range(n))

print(c)
```
